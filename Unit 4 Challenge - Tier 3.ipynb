**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:

```python
#I've already created a year column based on the prior exercise. Below I'll group by borough and by year 
#created a multi level list which I'll use to pull specific years and boroughs in my function below

GroupDF = Clean_Properties.groupby(['Borough','Year'])['Average Price'].mean()

#Get the total london housing average prices. Would be interesting to compare the individual boroughs to the group as a whole
TotalGroup = Clean_Properties.groupby('Year')['Average Price'].mean()

#Write function

#Use a Target Year and Lookback variable as inputs. This will make the function more flexible. I can put in '2018' and 10
#to look at the past 10 years from 2018 as the exercise instructed. Or any other combination. 2020, past 5 years
#2010 past 2 years, etc.

def Housing(TargetYear, Lookback):
    
    #define glogbal ChartTitle variable. Assign the variable within the function based on inputs and use it 
    #outside the function to name my chart
    global ChartTitle
    
    #establish unique list of boroughs from Clean_Properties datasets
    Borough = Clean_Properties['Borough'].unique()
    
    #establish empty dictionary which will be used to assign Borough and price change
    PropDict = {}
    
    #Calculate Total London Change
    TotalLondon = (TotalGroup[TargetYear] - TotalGroup[TargetYear - Lookback])/TotalGroup[TargetYear - Lookback] * 100
    
    #loop through each borough and specified year
    #calculate the percentage change
    
    for Bor in Borough:
       
        #calculate percentage change for each borough
        StartYear = TargetYear - Lookback
        CY_Price = GroupDF.loc[Bor,TargetYear]
        PY_Price = GroupDF.loc[Bor,StartYear]
        PctChange = (CY_Price-PY_Price)/PY_Price * 100
        
        #add to dictionary
        PropDict[Bor] = [PctChange,TotalLondon]
        
    #convert dictionary to dataframe, clean, and sort 
    df = pd.DataFrame(data=PropDict)
    df = df.transpose().reset_index()
    df.columns = ['Borough','Price Increase', 'Total London Increase']
    sort_df = df.sort_values('Price Increase', ascending = False)
    
    ChartTitle = "London Housing Price Change From "+ str(TargetYear - Lookback) + " to "+ str(TargetYear)
    return sort_df



#Call Housing function entering your target year and years to look back
#Based on instructions I'm entering 2018 as the target year and looking 10 years back

PriceInc = Housing(2018,10)

print(PriceInc)
#plot on graph
ax = PriceInc['Total London Increase'].plot(x = 'Borough', title = str(ChartTitle), figsize = (15,10), linestyle = "-", color = 'orange')
graph = PriceInc[['Borough','Price Increase']].plot(kind = 'bar',x = 'Borough', title = str(ChartTitle), figsize = (15,10), ax = ax)
plt.savefig('Housing Increases.png')
plt.show()



```

```python
#Look at price increases compared to Total London's increase
PriceInc['Relative Price Increase'] = PriceInc['Price Increase'] - PriceInc['Total London Increase']
graph = PriceInc[['Borough','Price Increase','Relative Price Increase']].plot(kind = 'barh', x = 'Borough', title = str(ChartTitle), figsize = (10,10))
plt.savefig(r'C:\Users\Nick\Google Drive\Springboard\London Housing\Relative Increases.png')
plt.show()

print(PriceInc)
PriceInc.to_excel('Price Increase Table.xls')
```

```python
#Additional Analysis
#Interested to see the borough that increased the most and the least by year

Clean_Properties.head()
df_pivot = Clean_Properties.pivot_table(index = 'Year', columns = 'Borough', values = 'Average Price', aggfunc = 'mean')
df_AnnualChange = df_pivot.pct_change()*100

dfIDMax = pd.DataFrame(df_AnnualChange.idxmax(axis = 1)).reset_index()
dfIDMax.columns = ['Year', 'Borough']

dfMax = pd.DataFrame(df_AnnualChange.max(axis = 1)).reset_index()
dfMax.columns = ['Year','Pct Change']

dfIDMin = pd.DataFrame(df_AnnualChange.idxmin(axis = 1)).reset_index()
dfIDMin.columns = ['Year', 'Borough']
dfMin = pd.DataFrame(df_AnnualChange.min(axis = 1)).reset_index()
dfMin.columns = ['Year', 'Pct_Change']

MaxChanges = pd.merge(dfIDMax, dfMax, on = 'Year')


MinChanges = pd.merge(dfIDMin, dfMin, on = 'Year')


AllChanges = pd.merge(MaxChanges,MinChanges, on='Year')
AllChanges.columns = ['Year','Best Perf Borough', 'Pct Change', 'Worst Perf Borough', 'Pct Change']
MaxCount = MaxChanges.groupby('Borough')['Year'].count()
MinCount = MinChanges.groupby('Borough')['Year'].count()

Counts = pd.merge(MaxCount, MinCount, on = 'Borough', how = 'outer').fillna(0)
Counts.columns = ['Frequency in Best List', 'Frequency in Worst List']

#AllChanges displays the best and worst performing boroughs by year
print(AllChanges)
AllChanges.to_excel('Best and Worst.xlsx')
#Counts displays how many times each borough showed up on the Best and Worst list
print(Counts)
Counts.to_excel('Counts of Best and Worst.xlsx')
```
